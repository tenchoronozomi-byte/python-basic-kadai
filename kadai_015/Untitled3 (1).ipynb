```python
class Human:
    def __init__(self, name, age):
        self.name = name      # 名前
        self.age = age        # 年齢

    def printinfo(self):
        print(f"名前: {self.name}, 年齢: {self.age}")


# --- Humanクラスのインスタンスを作成 ---
human1 = Human("太郎", 25)

# --- メソッドを呼び出して情報を表示 ---
human1.printinfo()
```
