```python
class Human:
    def __init__(self, name, age):
        self.name = name      # 名前
        self.age = age        # 年齢

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}さんは大人です")
        else:
            print(f"{self.name}さんは大人ではありません")


# --- Human のインスタンスを複数作成 ---
human_list = [
    Human("太郎", 25),
    Human("花子", 19),
    Human("次郎", 30),
]

# --- リストの要素分だけメソッドを呼び出す ---
for human in human_list:
    human.check_adult()
```
